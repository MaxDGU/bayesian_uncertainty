```python
import numpy as np
import matplotlib.pyplot as plt 
#FIBROBLASTS
file = open("valuesf.csv")
n = np.loadtxt(file, delimiter=",")
intervals = np.arange(1,52) #change to total # tests
DCN = n[:,0]
GSN = n[:,1]
PDGFRA = n[:,2]

plt.hist(DCN)
plt.title("DCN")
plt.ylabel("Test #")
plt.xlabel("Posterior")
plt.show()

plt.hist(GSN)
plt.title("GSN")
plt.ylabel("Test #")
plt.xlabel("Posterior")
plt.show()

plt.hist(PDGFRA)
plt.title("PDGFRA")
plt.ylabel("Test #")
plt.xlabel("Posterior")
plt.show()


```

```python
#SMOOTH MUSCLE CELLS
file = open("valuess.csv")
n = np.loadtxt(file, delimiter=",")
MYH11 = n[:,0]
TAGLN = n[:,1]
ACTA2 = n[:,2]

plt.hist(MYH11)
plt.title("MYH11")
plt.ylabel("Test #")
plt.xlabel("Posterior")
plt.show()
plt.hist(TAGLN)
plt.title("TAGLN")
plt.ylabel("Test #")
plt.xlabel("Posterior")
plt.show()
plt.hist(ACTA2)
plt.title("ACTA2")
plt.ylabel("Test #")
plt.xlabel("Posterior")
plt.show()


```
